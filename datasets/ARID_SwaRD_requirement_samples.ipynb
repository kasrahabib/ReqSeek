```python
import sys
sys.path.append("../datasets_with_data/ARID_supporting_scripts")

import mapper
import datasets
import numpy as np
import pandas as pd
```

```python
def display_dataset_counts(dataset_name, meta):
    c_names = np.unique(meta, return_counts = True)[0]
    c_counts = np.unique(meta, return_counts = True)[1]

    print(dataset_name)
    print(f"Sum instances: {c_counts.sum()}")
    
    for name, count in zip(c_names, c_counts):
        print(f"   {name} -> {count} instances")
```

# Random Sample Requirments: ARID

```python
# ARID full before removing instances from public sources we could not idenifty thier license, USED for ReqSeek Training
ARID_full = datasets.load_from_disk('../datasets_with_data/ARID_supporting_scripts/5_1_training_set')
ARID_full = datasets.concatenate_datasets([ARID_full['train'], ARID_full['test']])
isReqSrsTxt = mapper.map(ARID_full['signal_keyword'])
```

```python
display_dataset_counts('ARID with all public instances', isReqSrsTxt) 
```

```python
# ARID that has been made publicly abailable
ARID = datasets.load_from_disk('../datasets_with_data/ARID/')
```

```python
# Combine both train and text sets to draw random samples
ARID = datasets.concatenate_datasets([ARID['train'], ARID['test']])
```

```python
display_dataset_counts('ARID', ARID['ternary_cls']) 
```

```python
sub_columns = ['REQID_expanded', 'Project Name', 'Requirement Sentences', 'ternary_cls']
ARID_df = ARID.to_pandas()[sub_columns] # remove [[.....]], if you want to see all columns
```

```python
unselect = [
    'wikipedia',
    "A Dataset of Crowdsourced Smarthome Requirements with Creativity Ratings",
    "IlmSeven",
    "SecReq",
    "COEST",
    "PURE",
    "PromiseExp",
    "(RETRO).NET",
    "Requirements data sets (user stories)"]

ARID_df = ARID_df[~ARID_df["Project Name"].isin(unselect)].copy()
```

```python
ARID_sampled = (
    ARID_df
    .groupby("ternary_cls", group_keys = False)
    .apply(lambda x: x.sample(n = min(10, len(x)), random_state = 10))
    .style
    .hide(axis = "index")
    # general styling for all cells
    .set_properties(**{
        "text-align": "left",
        "direction": "ltr",
        "font-size": "12px",
        "line-height": "1.2",
        "padding": "4px",
        "vertical-align": "top"
    })
    .set_table_styles([
        {"selector": "table", "props": [
            ("table-layout", "fixed"),
            ("width", "100%"),
            ("border-collapse", "collapse")
        ]},
        {"selector": "th", "props": [
            ("text-align", "left"),
            ("direction", "ltr"),
            ("font-size", "12px"),
            ("padding", "4px")
        ]}
    ])
    # wrap ONLY the requirement text column
    .set_properties(
        subset = ['Requirement Sentences'],  
        **{
            "white-space": "normal",
            "word-break": "break-word",
            "max-width": "450px"
        }
    )
    .set_properties(
        subset = ['Project Name'],
        **{
            "white-space": "normal",
            "word-break": "break-word",
            "max-width": "200px"
        }
    )
)
```
