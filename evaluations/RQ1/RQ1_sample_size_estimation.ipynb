```python
# Common imports
import math
import numpy as np
import scipy.stats as stats
from statistics import NormalDist
```

# An Apriori Power Analysis for Friedman Test

```python
alpha = 0.05
target_power = 0.80
k_conditions = 4
cohens_f = 0.25
rm_corr = 0.50

# Approximate repeated-measures ANOVA adjustment
f_adjusted = cohens_f / math.sqrt(1 - rm_corr)

def compute_rm_anova_power(n, k, f_adj, alpha):
    df_effect = k - 1
    df_error = (n - 1) * (k - 1)
    lambda_nc = (f_adj ** 2) * n * k
    f_critical = stats.f.ppf(1 - alpha, df_effect, df_error)
    achieved_power = 1 - stats.ncf.cdf(f_critical, df_effect, df_error, lambda_nc)

    return achieved_power

n = 2

while compute_rm_anova_power(n, k_conditions, f_adjusted, alpha) < target_power:
    n += 1

print("Repeated-measures ANOVA approximation for Friedman test")
print(f"alpha: {alpha}")
print(f"target power: {target_power}")
print(f"Cohen's f: {cohens_f}")
print(f"assumed repeated-measures correlation: {rm_corr}")
print(f"adjusted Cohen's f: {f_adjusted:.4f}")
print(f"required sample size: {n}")
print(f"achieved power: {compute_rm_anova_power(n, k_conditions, f_adjusted, alpha):.3f}")
```

# Finite Population Sample Size

```python
def finite_population_sample_size(n_train: int, n_test: int, confidence: float = 0.95, margin_error: float = 0.10, p: float = 0.5) -> int:
    
    N = n_train + n_test
    
    if N <= 0: raise ValueError("N must be positive.")
    if not (0 < confidence < 1): raise ValueError("confidence must be between 0 and 1.")
    if not (0 < margin_error < 1): raise ValueError("margin_error must be between 0 and 1.")
    if not (0 < p < 1): raise ValueError("p must be between 0 and 1.")

    z = NormalDist().inv_cdf(1 - (1 - confidence) / 2)
    numerator = N * (z ** 2) * p * (1 - p)
    denominator = (margin_error ** 2) * (N - 1) + (z ** 2) * p * (1 - p)

    print(f"Sample Size: {math.ceil(numerator / denominator)}")
```

```python
finite_population_sample_size(n_train = 1916, n_test = 480)
```
